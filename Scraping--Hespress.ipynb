# Parsing URL

```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
from urllib.parse import urlsplit
from urllib.parse import urlparse
import urllib

site= 'https://www.hespress.com/politique'
hdr = {'User-Agent': 'Mozilla/5.0'}
req = Request(site,headers=hdr)
page = urlopen(req)
soup = BeautifulSoup(page)
```

# Get_all_URLs

```python
results=[]
date_1=[]
job_elems = soup.find_all('div', class_='col-12 col-md-7 col-lg-8 col-xl-9')
link=[]
for job_elem in job_elems:
    a = job_elem.find_all('a')
    for href in a:
        lien=href.get('href')
        link.append(lien)
```

# Date_processing

```python
import datetime

def arabic_converter(time):
    
    month_dict = {
        'كانون الثاني':'01',
        'يناير':'01',
        'فبراير':'02',
        'شباط':'02',
        'مارس':'03',
        'آذار':'03',
        'ابريل':'04',
        'نيسان':'04',
        'أبريل':'04',
        'مايو':'05',
        'أيار':'05',
        'ماي':'05',
        'يونيو':'06',
        'حزيران':'06',
        'يوليوز':'07',
        'تموز':'07',
        'يوليو':'07',
        'اغسطس':'08',
        'آب':'08',
        'أغسطس':'08',
        'غشث':'08',
        'شتنبر':'09',
        'سبتمبر':'09',
        'أيلول':'09',
        'أكتوبر':'10',
        'تشرين الأول':'10',
        'اكتوبر':'10',
        'نونبر':'11',
        'نوفمبر':'11',
        'تشرين الثاني':'11',
        'دجنبر':'12',
        'كانون الأولر':'12',
        'ديسمبر':'12'
            }
    
    trt0 = time.replace('بتاريخ ','')
    trt02 = trt0.replace('\t','')
    trt03 = trt02.replace(',','')
    trt1 = trt03.split(' ')
    date_list = trt1

    for char in date_list:
        if char in month_dict:
            month = month_dict[char]
        elif len(char) == 2 or len(char) == 1:
            day = char
        elif len(char) == 4:
            year = char

    date_time_str = year+'-'+month+'-'+day
    result = datetime.datetime.strptime(date_time_str, '%Y-%m-%d')
    return(result)
```

# Structuring and Scraping of articles 


```python
list_pricipale=[]
for i in range(0,len(link)):
    para=''
    list_r=[]
    site= link[i]
    hdr = {'User-Agent': 'Mozilla/5.0'}
    req = Request(site,headers=hdr)
    page = urlopen(req)
    soup2 = BeautifulSoup(page)
    content = soup2.find_all('section', class_='apost')
    for cont in content:
        titre= cont.find_all('h1', class_='post-title')
        Date= cont.find_all('span', class_='date-post')
        
    list_r.append(link[i])
    for t in titre:
        titre2 = t.text
        titre2 = titre2.replace('\n','')
        titre2 =titre2.replace("\t",'')
        list_r.append(titre2)
    for h in Date:
        date = h.text
        date = date.replace('\n','')
        date =date.replace("\t",'')
        date =date.replace("\r",'')
        date =date.replace("          ",'')
        list_r.append(date)
    parag= cont.find_all('p')
    #print('Contenu:--------------------------------------------------------')
    for p in parag:
        contenu = p.text
        para=para+' '+contenu
       #print(contenu)
    para = para.replace('\n','')
    para =para.replace("\t",'')
    list_r.append(para)
    list_pricipale.append(list_r)
#         results.append([Date,lien,titre2,contenu])
```
